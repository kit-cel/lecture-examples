```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definiere Parameter
max_time = 10
time_step_size = 0.1
no_samples_per_time_step = 100000
time_steps = np.arange(0, max_time, time_step_size)
mean = time_steps
variance = np.ones((time_steps.size,))

# Sample
samples = np.random.normal(loc=mean,scale=variance, size=(no_samples_per_time_step,1,time_steps.size))
```

```python
# Plote die Verteilung über der Zeit
plt.figure()
time_2d = np.tile(time_steps, (no_samples_per_time_step, 1))
hist = plt.hist2d(time_2d.flatten(), samples[:,0,:].flatten(), bins=100, cmap='jet')
plt.colorbar(label='Häufigkeit')
plt.xlabel('Zeit')
plt.ylabel('Beobachtung')
plt.title('Verteilung über der Zeit')
```

```python
# Plote einzelne Realisierungen
realizations = np.array([0, 6, 17])

plt.figure()
plt.plot(np.transpose(np.squeeze(samples[realizations,:,:])))
plt.xlabel("Zeit")
plt.ylabel("Beobachtung")
plt.title("Realisierungen/Pfade")
plt.legend(["Pfad %d" % realizations[0],"Pfad %d" % realizations[1], "Pfad %d" % realizations[2]])
```

```python
analysis_timesteps = np.array([0, 50, 99])
plt.figure()
counts, bins = np.histogram(np.squeeze(samples[:,:,analysis_timesteps[0]]), bins = 100, density=True)
plt.stairs(counts, bins)
counts, bins = np.histogram(np.squeeze(samples[:,:,analysis_timesteps[1]]), bins = 100, density=True)
plt.stairs(counts, bins)
counts, bins = np.histogram(np.squeeze(samples[:,:,analysis_timesteps[2]]), bins = 100, density=True)
plt.stairs(counts, bins)
plt.xlabel("Beobachtung")
plt.ylabel("f(n)")
plt.title("Wahrscheinlichkeitsverteilungen")
plt.legend(["Zeitpunkt %0.1f" % time_steps[analysis_timesteps[0]], "Zeitpunkt %0.1f" % time_steps[analysis_timesteps[1]], "Zeitpunkt %0.1f" % time_steps[analysis_timesteps[2]]])     
```
