# Content and Objective

+ Illustrate effects of estimation in a Bernoulli experiment
+ Bernoulli experiments are observed and resulting estimations are shown

# Importing

```python
# importing
import numpy as np

from scipy import stats, special


import matplotlib.pyplot as plt
import matplotlib

# showing figures inline
%matplotlib inline
```

```python
# plotting options 
font = {'size'   : 20}
plt.rc('font', **font)
plt.rc('text', usetex=True)

matplotlib.rc('figure', figsize=(18, 6) )
```

# Parameters

```python
# P(1)
p = .6

# observation length for estimation
# NOTE: corresponding to N out of the lecture
N_1 = 20 
N_2 = 200

# number of trials for estimation expection of the estimator
# NOTE: not relativ to observation length
N_trials = int( 1e3 )
```

# Simulations

```python
# sample random values
# Note: 2-dim arrays are sampled in which
#       first dimension describes N realizations used for determining the estimator
#       second dimension realizes N_trials used for averaging w.r.t. different estimation outcomes

X_samples_1 = np.random.choice( [0,1], size=(N_1, N_trials), p=[1-p, p] )
X_samples_2 = np.random.choice( [0,1], size=(N_2, N_trials), p=[1-p, p] )

# determine average w.r.t. to first dimension = axis 0 
# Note: there are still N_trials realizations, which is why an array will be created
X_average_1 = np.average( X_samples_1, axis=0 )
X_average_2 = np.average( X_samples_2, axis=0 )
```

### Show histogram of estimated values

```python
plt.figure()

plt.subplot(121)
plt.hist( X_average_1, bins=50, density=1, label='$h_{N}[n], N=$'+str(N_1) )
plt.plot( [ p, p], [0, 10], '-x', color='red', ms=16, linewidth=2.0, label='true value')

plt.xlim( (0,1) )
plt.xlabel('$\widehat{p}$')
plt.grid( 1 )
plt.legend( )

plt.subplot(122)
plt.hist( X_average_2, bins=50, density=1, label='$h_{N}[n], N=$'+str(N_2) )
plt.plot( [ p, p], [0, 10], '-x', color='red', ms=16, linewidth=2.0, label='true value')
plt.xlim( (0,1) )
plt.xlabel('$\widehat{p}$')
plt.grid( 1 )
plt.legend( )
```

### Check if estimator is unbiased

```python
# determine E( estimator ) 
E_T = np.average( X_average_1 )
V_T = np.var( X_average_1 )

print(' Expectation of estimator - theory: \t{}'.format(p) )
print(' Average of estimators: \t\t{}\n\n'.format( E_T ) )

print(' Variance of estimator -  theory: \t{:2.7f}'.format( p * ( 1 - p ) / N_1 ) )
print(' Variance of estimators: \t\t{:2.7f}'.format( V_T ) )
```
