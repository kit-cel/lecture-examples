+ Check approximation of the sum of Cauchy distributed random variables by the central limit theorem

# Import

```python
# importing
import numpy as np

from scipy import stats, special

import matplotlib.pyplot as plt
import matplotlib

# showing figures inline
%matplotlib inline
```

```python
# plotting options 
font = {'size'   : 20}
plt.rc('font', **font)
plt.rc('text', usetex=True)

matplotlib.rc('figure', figsize=(18, 6) )
```

# Show Cauchy PDF

```python
x = np.arange( -10, 10, .01)
f = stats.cauchy.pdf( x )
plt.plot( x, f, label='$f_X(x)$' )
plt.xlabel('$x$')
plt.grid(1)
plt.legend()
```

# Adding up Cauchy Distributed Random Variables

```python
# number of RVs to be added
N = int( 1e3 )

# simulation
N_trials = int( 1e6 )

# get Cauchy distributed values
X = np.zeros( (N, N_trials) )
for _n in range( N ):
    X[ _n, : ] = stats.cauchy.rvs( size=N_trials )

# perform sample average
Z = 1/N * np.sum( X, axis = 0 )
```

```python
# plotting    
[ hist, edges] = np.histogram( Z, 1000, density=True )

plt.semilogy( edges[1:], hist )
plt.xlabel('$n$')
plt.ylabel('$H_{{{}}}(n)$'.format(N_trials)) 
plt.grid( True )
plt.legend( loc = 'upper right' )
```
