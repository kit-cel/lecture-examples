# Import

```python
# importing
import numpy as np

from scipy import stats, special
from decimal import Decimal

import matplotlib.pyplot as plt
import matplotlib

# showing figures inline
%matplotlib inline
```

```python
# plotting options 
font = {'size'   : 20}
plt.rc('font', **font)
plt.rc('text', usetex=True)

matplotlib.rc('figure', figsize=(18, 6) )
```

# Generate Sequences of Cauchy and Show their Mean

```python
# number of sequences
N_seq = int( 10 )

# number of samples per sequences
N_samples = int( 1e7 )

# initialize sequences 
X = np.zeros( (N_seq, N_samples ) )
```

```python
# sample sequences
sample_indices = 1 + np.arange( N_samples )
for _n in range( N_seq ):
    samples = stats.cauchy.rvs( size =  N_samples )

    X[ _n, : ] = np.cumsum( samples ) / sample_indices 
```

```python
plt.figure()
for _n in range( N_seq ):
    plt.plot( X[ _n, : ] )
plt.grid(1)
plt.ylim( (-100,100) )
```
