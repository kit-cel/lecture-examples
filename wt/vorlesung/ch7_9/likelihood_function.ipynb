# Content and Objective

+ Show likelihood function of binomial distribution
    + For several values of "hits", according pmf is calculated
    + y axis may be depicted logarithmically or linearly, the latter leading to loss of graphical information due to very different values

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

# Parameters

```python
# possible parameters vartheta 
delta = .01
vartheta = np.arange( 0, 1 + delta, delta )

# number of trials
N = 100

# get vector for counting hits
t = np.arange( 50, 110, 10 )
```

# Get Values

```python
# theoretical values
p = np.zeros( ( len( t ), len( vartheta ) ) )
for ind_t, val_t in enumerate( t ):
    p[ ind_t, : ] = vartheta ** val_t * ( 1 - vartheta ) ** (N - val_t)
```

```python
plt.figure()
plt.figure( figsize=( 12, 4 ) )

for ind_t, val_t in enumerate( t ):
    plt.semilogy( vartheta, p[ind_t, :], label='$p_{\\vartheta}(\mathbf{x}), t =$'+str(val_t) )
plt.xlabel('$\\vartheta$')
plt.grid()
plt.legend()
```
