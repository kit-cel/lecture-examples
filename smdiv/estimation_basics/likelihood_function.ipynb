# Content and Objective

+ Show likelihood function of binomial distribution
    + For several values of "hits", according pmf is calculated
    + y axis may be depicted logarithmically or linearly, the latter leading to loss of graphical information due to very different values

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
delta = .01
theta = np.arange( 0, 1 + delta, delta )
```

```python
N = 100
t = np.arange( 50, 110, 10 )

p = np.zeros( ( len(t), len(theta) ) )
for ind_t, val_t in enumerate( t ):
    p[ ind_t, : ] = theta ** val_t * (1-theta) ** (N - val_t)
```

```python
plt.figure()
plt.figure( figsize=( 12, 4 ) )

for ind_t, val_t in enumerate( t ):
    plt.semilogy(theta, p[ind_t, :], label='$p_{\\vartheta}(\mathbf{x}), t =$'+str(val_t) )
plt.xlabel('$\\vartheta$')
plt.grid()
plt.legend()


```
