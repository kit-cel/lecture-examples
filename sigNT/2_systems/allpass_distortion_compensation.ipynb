# Content and Objective

+ Show behavior of an allpass for distortion compensation

# Importing and Plotting Options

```python
import numpy as np

from scipy import signal

import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# plotting options 
font = {'size'   : 20}
plt.rc('font', **font)
plt.rc('text', usetex=True)

matplotlib.rc('figure', figsize=(24, 10) )
```

# Show Example

### defining frequency regime

```python
delta_Omega = .01
Omega = np.arange( 0, 2 * np.pi, delta_Omega )
```

### defining frequency response

```python
# directly give H(Omega)
H = np.exp( 1j * Omega**2 )
```

### plotting frequency response

```python
ax = plt.subplot(131)

plt.plot( Omega, np.abs( H ), label='$|H(\\Omega)|$' )

plt.grid( True )
plt.xlabel('$\\Omega$')
plt.ylabel('$|H(\\Omega)|$')
plt.legend(loc='upper right')

plt.subplot(132)
plt.plot( Omega, np.angle( H ), label='$\\angle H(\\Omega)$' )

plt.grid( True )
plt.xlabel('$\\Omega$')
plt.ylabel('$\\angle \\Omega$')
plt.legend(loc='upper right')


plt.subplot(133)
plt.plot( Omega, - np.diff( np.unwrap( np.angle( H ) ), prepend=0 ), label='$\\tau_\mathrm{g} (\\Omega)$' )

plt.grid( True )
plt.xlabel('$\\Omega$')
plt.ylabel('$\\tau_\mathrm{g} (\\Omega)$')
plt.legend(loc='upper right')
plt.show()
```
